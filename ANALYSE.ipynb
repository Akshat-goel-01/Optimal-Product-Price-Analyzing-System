```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
dt = datetime.datetime.now()
date = dt.strftime("%d_%m_%Y")
date
```

```python
master_df = pd.read_excel('Analyse/master1.xlsx')
filter_df = pd.read_excel('Analyse/filtered_output.xlsx')
```

```python
header = list(filter_df.columns)
header[0] = date + "_" + str(header[0])
header[1] = date + "_" + str(header[1])
header[2] = date + "_" + str(header[2])
header[3] = date + "_" + str(header[3])
header[4] = date + "_" + str(header[4])
header[5] = date + "_" + str(header[5])
header[6] = date + "_" + str(header[6])

dict = {'Product': header[0],
        'Amazon_price': header[1],
        'Amazon_rating': header[2],
        'Amazon_rating_count': header[3],
        'Flipkart_price': header[4],
        'Flipkart_rating': header[5],
        'Flipkart_rating_count': header[6]}
 
filter_df.rename(columns=dict, inplace=True)
```

```python
master_df = pd.concat([master_df, filter_df], axis=1)
master_df.to_excel('Analyse/master.xlsx', index = False)
```
